**Imbalanced Dataset**
A dataset with an unequal class distribution

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/credit_data.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python

df.fillna(0, inplace=True)  # or use a different value/method

#distribution of two classes
df['Class'].value_counts()
```

This is highly imbalanced dataset
0- Legit
1- Fradulent

```python
df['Class'] = df['Class'].astype(int)
df.head()
```

```python
legit=df[df.Class==0]
fraud=df[df.Class==1]
print(legit.shape)
print(fraud.shape)
```

Undersampling

---
Undersampling means reducing the number of examples in the majority class so it matches (or is closer to) the number of examples in the minority class.

```python
# build a sample dataset containing the same distribution of legit and fraud transaction
```

```python
legit_sample=legit.sample(n=94)
```

```python
legit_sample.shape
```

```python
#Concatenate the dataframe
```

```python
new_df=pd.concat([legit_sample,fraud],axis=0)
new_df.head()
```

```python
new_df['Class'].value_counts()
```
